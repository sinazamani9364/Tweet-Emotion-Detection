```python
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
import json
```

```python
def read_data_file(file_name):
  data = pd.read_csv(file_name, usecols=['Tweet', 'Label'])

  return data.to_dict()

tweets = read_data_file('labeled_data.csv')
```

```python
print(len(tweets['Tweet']))
print(len(tweets['Label']))
```

```python
print(tweets)
```

```python
def organize_data(data):
  tweets = {}
  for idx in range(len(data['Tweet'])):
    tweets[idx] = {}
    tweets[idx]['text'] = data['Tweet'][idx]
    tweets[idx]['label'] = data['Label'][idx]
    tweets[idx]['sentence_broken'] = []
    tweets[idx]['word_broken'] = []

  return tweets

tweets = organize_data(tweets)
```

```python
print(tweets)
```

```python
def sentence_tokenizer(text):
  sentences = sent_tokenize(text)
  return sentences
```

```python
def word_tokenizer(sentence):
  words = word_tokenize(sentence)
  return words
```

```python
def tokenize(data):
  for idx in range(len(data)):
      text = data[idx]['text']
      data[idx]['sentence_broken'].extend(sentence_tokenizer(text))
      data[idx]['word_broken'].extend(word_tokenizer(text))

  return data

tweets = tokenize(tweets)
```

```python
sadness_tweets = {}
happiness_tweets = {}
fear_tweets = {}
anger_tweets = {}
surprise_tweets = {}
disgust_tweets = {}
neutral_tweets = {}

for idx in range(len(tweets)):
  if tweets[idx]['label'] == 'sadness':
    sadness_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'happiness':
    happiness_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'fear':
    fear_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'anger':
    anger_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'surprise':
    surprise_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'disgust':
    disgust_tweets[idx] = tweets[idx]
  elif tweets[idx]['label'] == 'neutral':
    neutral_tweets[idx] = tweets[idx]
```

```python
with open("sadness_tweets.json", "w") as json_file:
  json.dump(sadness_tweets, json_file)

with open("happiness_tweets.json", "w") as json_file:
  json.dump(happiness_tweets, json_file)

with open("fear_tweets.json", "w") as json_file:
  json.dump(fear_tweets, json_file)

with open("anger_tweets.json", "w") as json_file:
  json.dump(anger_tweets, json_file)

with open("surprise_tweets.json", "w") as json_file:
  json.dump(surprise_tweets, json_file)

with open("disgust_tweets.json", "w") as json_file:
  json.dump(disgust_tweets, json_file)

with open("neutral_tweets.json", "w") as json_file:
  json.dump(neutral_tweets, json_file)
```
